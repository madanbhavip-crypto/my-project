## Feature Extraction
This code extracts meaningful features from `DATASET_FILE` and stores them in `FEATURES_FILE`. 

```python
import numpy as np
import pandas as pd

DATASET_FILE  = 'Gesture_Dataset.csv'
FEATURES_FILE = 'features.csv'
```

### Feature Extraction Function
This cell defines a new function to be able to extract the features from the input data.
It normalises the data and extracts the mean, standard deviation, maximum, minimum and RMS values from each row. It will append the same data to the end of each new column in the `features.csv` file.

```python
def extract_features(rows):
    
    rows = (rows - rows.mean(axis=0)) / (rows.std(axis=0) + 1e-8)
    
    features = []
    for col in range(6):
        s = rows[:, col]
        features += [np.mean(s), np.std(s), np.min(s), np.max(s),
                     np.max(s)-np.min(s), np.sqrt(np.mean(s**2))]
    
    for col in range(6):
        s = rows[:, col]
        features.append(np.mean(np.abs(np.diff(s))))
    
    return features  
```

### Feature Extraction

Here, we load the `DATASET_FILE` and start feature extraction. All of the data is saved in `FEATURES_FILE`. At the end, we display the table of `FEATURES_FILE`.

```python
df = pd.read_csv(DATASET_FILE)
print(f"Raw data: {len(df)} rows, {df['sample_id'].nunique()} gestures")

X, y = [], []

for sid, group in df.groupby('sample_id'):
    rows = group[['ax','ay','az','gx','gy','gz']].values.astype(float)
    X.append(extract_features(rows))
    y.append(group['label'].iloc[0])

features_df = pd.DataFrame(X)
features_df.insert(0, 'label', y)

features_df.to_csv(FEATURES_FILE, index=False)
print(f"Saved {len(features_df)} feature vectors → '{FEATURES_FILE}'")
features_df.head()
```
